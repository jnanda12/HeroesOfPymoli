## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
number_unique = purchase_data['Item ID'].nunique()
total_revenue = '${:,.2f}'.format(purchase_data['Price'].sum())
average_price = '${:,.2f}'.format(purchase_data['Price'].mean())
number_purchases = purchase_data['Purchase ID'].nunique()

summary_df = pd.DataFrame({'Number of Unique Items':number_unique, 'Average Price': average_price,\
                           'Number of Purchases':number_purchases, 'Total Revenue':total_revenue}, index=[0])
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_data.head()
```

```python
#purchase_data.loc[purchase_data['Gender']=='Male'].count()
player_counts = pd.DataFrame(purchase_data['Gender'].value_counts())
total_players = player_counts['Gender'].sum()
player_counts['Percentage of Players'] = player_counts.apply(lambda x: ((x/total_players)*100))
player_counts['Percentage of Players'] = pd.Series(["{0:.2f}%".format(val) for val in player_counts['Percentage of Players']], index = player_counts.index)
player_counts.head()
```

```python
female = purchase_data.loc[purchase_data['Gender']=='Female']
male = purchase_data.loc[purchase_data['Gender']=='Male']
other = purchase_data.loc[purchase_data['Gender']=='Other / Non-Disclosed']
male_avg = male['Price'].mean()
female_avg = female['Price'].mean()
other_avg = other['Price'].mean()
male_total = male['Price'].sum()
female_total = female['Price'].sum()
other_total = other['Price'].sum()
male_group = male.groupby('SN')
male_avgpp = sum(male_group['Price'].sum())/sum(male_group['SN'].nunique())
female_group = female.groupby('SN')
female_avgpp = sum(female_group['Price'].sum())/sum(female_group['SN'].nunique())
other_group = other.groupby('SN')
other_avgpp = sum(other_group['Price'].sum())/sum(other_group['SN'].nunique())

player_counts['Average Purchase Price'] = pd.Series(["${0:.2f}".format(male_avg), "${0:.2f}".format(female_avg), "${0:.2f}".format(other_avg)], index=player_counts.index)
player_counts['Total Purchase Value'] = pd.Series(["${0:.2f}".format(male_total), "${0:.2f}".format(female_total), "${0:.2f}".format(other_total)], index=player_counts.index)
player_counts['Average Total Purchase Per Person'] = pd.Series(["${0:.2f}".format(male_avgpp), "${0:.2f}".format(female_avgpp), "${0:.2f}".format(other_avgpp)], index=player_counts.index)

player_counts

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
binss = [0, 9, 14, 19, 24, 29, 34, 39, 1000]
labelss = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
bin_df = pd.cut(purchase_data['Age'], bins =binss, labels = labelss)

bin_df = pd.DataFrame({'Total Count' : bin_df.value_counts()})
bin_df['Percentage of Players'] = bin_df['Total Count'].apply(lambda x: (x/bin_df['Total Count'].sum()*100))
bin_df['Percentage of Players'] = pd.Series(["{0:.2f}%".format(val) for val in bin_df['Percentage of Players']], index = bin_df.index)
bin_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
binss = [0, 9, 14, 19, 24, 29, 34, 39, 1000]
labelss = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], bins =binss, labels = labelss)
age_group = purchase_data.groupby('Age Group')
tot_purchase = age_group['Price'].sum()
avg_purchase = age_group['Price'].mean()
avgpp = age_group['Price'].sum()/age_group['SN'].nunique()

bin_df = pd.DataFrame({'Purchase Count' : age_group['Purchase ID'].count(), 'Average Purchase Price':(avg_purchase), \
                      'Total Purchase Value':(tot_purchase), 'Avg Total Purchase per Person':(avgpp)})
for i in bin_df.columns[1:]:
    bin_df[i] = pd.Series(["${0:.2f}".format(val) for val in bin_df[i]], index = bin_df.index)
bin_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
```

```python
pop_df = purchase_data[['Item ID', 'Item Name', 'Price']]
pop_df = pop_df.groupby(['Item ID', 'Item Name'])
tot_pv = pop_df['Price'].sum()
pur_count = pop_df['Item Name'].count()
price = pop_df['Price'].mean()

most_pop = pd.DataFrame({'Purchase Count': pur_count, 'Item Price': price, 'Total Purchase Value': tot_pv})

for i in most_pop.columns[1:]:
    most_pop[i] = pd.Series(["${0:.2f}".format(val) for val in most_pop[i]], index = most_pop.index)
most_pop.sort_values('Purchase Count', ascending = False).head()
```
